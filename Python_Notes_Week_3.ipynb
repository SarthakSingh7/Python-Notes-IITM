##**Poblem 1: Find the factorial of the given number**

````python
# Solution from the last tutorial

#Problem 1: Find the factorial of the given number

n = int(input('enter a number: '))

fact = 1    # fact denotes factorial

if (n<0):
  print('Not defined') 

else: 
  while(n>0):
    fact = fact * n
    n = n - 1
  print(fact)
````

We will try to convert the **while** loop which was used in the previous tutorial to its equivalent **for** loop

````python
#Problem 1: Find the factorial of the given number

num = int(input('Enter a number: '))
fact = 1
if(num < 0):
    print('Not defined')
else:
    for i in range(num, 1, -1):
        fact = fact * i
    print(fact)
````

## Note 1 
Here, the range function has three parameters first, the number that is the starting point, the ending point and the step count, that means the value of number will start from the given input and then it will do -1 for every iteration till it reaches to 1 and the updated value will be stored in variable i which we are using to multiply with variable factorial.
Once the entire for loop execution is finished, we are printing the result which is stored in variable fact.

##Note 2
As you can see, both programs using while loop as well as for loop looks identical,

but they have a little bit of difference, as a programmer we have to decide which loop
is more ideal for this particular example.

Now, the question is how to decide that what are the factors which we should consider before
deciding whether to use while loop or for loop?

So, **the most important factor we should consider is whether do we know the range of the loop**
or not, as in are we aware in advance that the loop is going to run for some n number
of times, ***if we are certain that the given loop is going to execute for let us say 10
times or 12 times then we should opt for for loop.***

Whereas, if we do not know how many number of times the particular loop is going to execute,
we should opt for **while loop.**

*This is important because whenever we use for loop we have to provide this range function*
and inside that range function, we should also mention the starting and ending point of the loop.

In order to provide these counts, we should know how many number of times the loop will
execute.


*Hence, **for** is more suitable when we know the total number of iterations required in that
particular program, whereas **while** is more generic way of execution.*



##Note 3 regarding the problem statement 1

As we can see, this is a problem statement where we are trying to find the factorial
of a number and in such a case, the number of iterations are known to us.
Hence, **for** loop is more suited option than a **while** loop.

##**Poblem 4: Find whether the entered number is palindrome or not**

````python
# solution to problem 4 ( from previous tutorial)

n = int(input("Enter a number: "))
absnum = abs(n)
rev = absnum % 10
absnum = absnum // 10 
while (absnum > 0):
    r = absnum % 10
    absnum = absnum // 10 
    rev = rev * 10 + r
if (n < 0):
  rev = rev - 2 * rev

if (n == rev):               #if - while block for positive numbers in rev order
  print('Palindrome')

else:                     #else - while block for negative numbers in rev order
  print('Not a palindrome')
````

**Working of the code above**

We are just reversing the digits in the number and if they are equal to the input we call it a *palindrome*, if not then we called it *not a palindrome*.
  

in this case we cannot predict how many ilterations are required in order to reverse that number
therefore it is not possible to determine a range for the **for** loop.
so, we will try to convert the number into its equivalent for loop and use the **for each** feature of the for loop

````python
'''Problem 4: Find whether the entered number is palindrome or not'''
num = int(input('Enter a number: '))
absStrNum = str(abs(num))
rev = ''
for c in absStrNum:
    rev = c + rev
if(num < 0):
    rev = '-' + rev
if(num == int(rev)):
    print('Palindrome')
else:
    print('Not a palindrome')
````

**The working and the logic of the code above:**

it is converting the input number into a string
representation, then using for each functionality of for loop, we are reversing the characters in the given representation of the string number, then we will check whether the original
number is less than 0 or not.

If that is the case, we will append a negative sign before the number, then we will compare
the original number with the reverse number and if they are equal, then we will print
palindrome else not a palindrome.


**Note:**

This particular code using for loop is also not a ideal representation
of this particular problem statement.
Hence, it is better to write this program using a while loop.

````python
'''Problem 1: Find all prime numbers less than the entered number'''

num = int(input('Enter a number: '))

if(num > 2):
    print(2, end = ' ')  # 2 is the only even and prime number

for i in range(3, num):
    flag = False
    for j in range(2, i):
        if(i % j == 0):
            flag = False
            break
        else:
            flag = True
    if(flag):
        print(i, end = ' ')
````

````python
# improved

'''Problem 1: Find all prime numbers less than the entered number'''

num = int(input('Enter a number: '))

if(num == 1 or num == 2):
  print("No output")

if(num <= 0):
  print("prime numbers are always positive and greater than 1")

if(num > 2):
    print(2, end = ' ')  # 2 is the only even and prime number

for i in range(3, num):
    flag = False
    for j in range(2, i):
        if(i % j == 0):
            flag = False
            break
        else:
            flag = True
    if(flag):
        print(i, end = ' ')
````

##Note
Prime numbers are the **positive integers** having only two factors, **1 and the integer itself**. For example, factors of 6 are 1, 2, 3 and 6, which are four factors in total. We can also say that the prime numbers are the numbers, which are only divisible by 1 and the number itself. Another way of defining it is a positive number or integer, which is not a product of any other two positive integers. **Always remember that 1 is neither prime nor composite.** Thus, **all prime numbers are greater than 1. Also, we can say that except 1, remaining numbers are classified as prime and composite numbers.** 

There is no defined formula to find if a number is prime or not (except to a certain range), apart from finding its factors. However, the numbers such as 7 have only two factors, i.e. 1 and 7. Hence, 7 is a prime number, but 6 is not; instead, it is a composite number.

````python
'''Problem 2: Find the total profit/loss of each trader working in
 a trading firm. Information regarding number of
  traders and number of transactions is unknown'''

empID = input('Enter employee ID: ')
while(empID != '-1'):
    trade = int(input('Enter the trade amount: '))
    profit_loss = 0
    while(trade != 0):
        profit_loss = profit_loss + trade
        trade = int(input('Enter the trade amount: '))
    print(f'Profit/loss for employee {empID} is {profit_loss}') #formatted printing

    empID = input('\nEnter employee ID: ') 
````

````python
'''Problem 4: Find the length of longest word from the set of words entered by the user'''
word = input('Enter a word: ')
maxLen = 0
while(word != '-1'):
    count = 0
    for letter in word:
        count = count + 1
    if(count > maxLen):
        maxLen = count
    word = input('Enter a word: ')
print('The length of longest word is %s' %maxLen)
````

##**Break**

Break statement in Python is used to bring the control out of the loop when some external condition is triggered. Break statement is put inside the loop body (generally after if condition).

##**Break**

###**Problem statement:** accept the email id from the user and then just print the user id part from that email.

example: 

input: xyz123@iitm.in

expected output: xyz123

````python
# given solution 

email = input("Enter email id: ")
for i in email:
  if(i == '@'):
    break
  print(i, end = "")
````

##Note
The **break** statement is like the **exit loop statement** 
in the above example it will stop the for loop if the condition of the **"if"** statement is met, as soon as the for loop stops the computer will directly jump to the line 8 which is blank in this case.

##**Continue statement**

similar problem statement 

###**Problem statement:** accept the email id from the user and then print the user name and domain name part from that email.

example: 

input: xyz123@iitm.in

expected output: 

user name: xyz123

domain name: iitm.in

note: just '@' is not present in the output

````python
# given solution 

email = input("Enter email id: ")
for i in email:
  if(i == '@'):
    print("") 
    continue
  print(i, end = "")
````

##Note
###Definition of **continue statement** from Geeks for geeks

Continue statement is a loop control statement that forces to execute the next iteration of the loop while skipping the rest of the code inside the loop for the current iteration only i.e. when the continue statement is executed in the loop, the code inside the loop following the continue statement will be skipped for the current iteration and the next iteration of the loop will begin.
