````python
# shift to any number of times, earlier it was only one shift
# this is popularily called as the Ceasar Cypher in Cryptography

alpha = 'abcdefghijklmnopqrstuvwxyz'

s= str(input('Enter a 6 letter word all in small letters:' ))


t='' 

i = 0
K= int(input('how many shifts in each letter? '))

#in the code below replace 1 by k
#t= t+(alpha[(((alpha.index(s[i]))+1)%26)]) 

#new code: t= t+(alpha[(((alpha.index(s[i]))+K)%26)]) 

t= t+(alpha[(((alpha.index(s[i]))+K)%26)]) 
t= t+(alpha[(((alpha.index(s[i+1]))+K)%26)])
t= t+(alpha[(((alpha.index(s[i+2]))+K)%26)])
t= t+(alpha[(((alpha.index(s[i+3]))+K)%26)]) # this is concatenation
t= t+(alpha[(((alpha.index(s[i+4]))+K)%26)])
t= t+(alpha[(((alpha.index(s[i+5]))+K)%26)])  


print(t)


# test case 1: if the input is 'bharat' and 1 shift is done then 
#the expected output is: 
# 'cibsbu' (quotes not included)

# test case 2: if the input is 'bharat' and 5 shifts is done then 
#the expected output is: 
# 'gmfwfy' (quotes not included)
````
